```python
import random

class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.hp > 0

    def take_damage(self, damage):
        actual_damage = damage - self.defense
        if actual_damage > 0:
            self.hp -= actual_damage
            print(f"{self.name}ì´(ê°€) {actual_damage}ì˜ ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ìŠµë‹ˆë‹¤. (ë‚¨ì€ ì²´ë ¥: {self.hp})")
        else:
            print(f"{self.name}ì€(ëŠ”) ë°ë¯¸ì§€ë¥¼ ë°›ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"{self.name}ì´(ê°€) {target.name}ì„(ë¥¼) ê³µê²©í•©ë‹ˆë‹¤! (ë°ë¯¸ì§€: {damage})")
        target.take_damage(damage)

class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}ì´(ê°€) ê²½í—˜ì¹˜ {exp}ì„(ë¥¼) íšë“í–ˆìŠµë‹ˆë‹¤. (í˜„ì¬ ê²½í—˜ì¹˜: {self.experience})")

    def level_up(self):
        while self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f"ğŸ‰ ë ˆë²¨ ì—…! {self.name}ì˜ ë ˆë²¨ì´ {self.level}ì´ ë˜ì—ˆìŠµë‹ˆë‹¤.")

class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)

def battle(player, monster):
    print(f"\nâš”ï¸ ì „íˆ¬ ì‹œì‘! {player.name} VS {monster.name} (Lv.{monster.level}) âš”ï¸")

    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():
        exp = monster.level * 20
        player.gain_experience(exp)
        player.level_up()
        print("ğŸ† ì „íˆ¬ ìŠ¹ë¦¬!")
    else:
        print("ğŸ’€ ì „íˆ¬ íŒ¨ë°°..")
```
