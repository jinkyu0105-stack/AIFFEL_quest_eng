```python
from datetime import datetime

class Kiosk:
    def __init__(self):
        self.menu = [
            ['americano', 2000],
            ['latte', 3000],
            ['mocha', 3000],
            ['yuza_tea', 2500],
            ['green_tea', 2500],
            ['choco_latte', 3000]
        ]
        
    # 메뉴 출력 메서드
    def menu_print(self):
        for idx, item in enumerate(self.menu, start=1):
            print(f'{idx} {item[0]} : {item[1]}원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = [] # 누적 주문 목록
        self.order_price = [] # 누적 주문 금액
        # 음료 번호 입력 받음
        while True:
            try:
                n = int(input("\n음료 번호를 입력하세요 (0: 주문 완료) :"))
            except ValueError:
                print('\n정수를 입력하세요')
                continue
            # 0입력시, 주문 완료.
            if n == 0:
                break

            # 음료 번호를 맞게 입력 했을 때
            if 1 <= n <= len(self.menu):
                
                # Hot, Iced 선택
                while True:
                    try:
                        t = int(input('HOT = 1 / ICED = 2 :'))
                    except ValueError:
                        print('정수를 입력하세요\n')
                        continue
                    if t == 1:
                        temp = "HOT"
                        break
                    elif t == 2:
                        temp = "ICED"
                        break
                    else:
                        print('HOT = 1 / ICED = 2 중 하나를 입력하세요\n')

                #메뉴에 누적시킴
                name, price = self.menu[n-1]
                self.order_menu.append(temp + ' ' + name)
                self.order_price.append(price)
                
                # 추가한 메뉴, 합계 표시
                print(f'추가된 주문: {temp} {name} : {price} 원')
                print(f'합계: {sum(self.order_price)}')

            # 없는 번호를 입력했을 때
            else:
                print('없는 메뉴입니다. 다시 주문해 주세요.')
        
        print("\n 주문 내역")
        for i in range(len(self.order_menu)):
            print(f'{self.order_menu[i]} : {self.order_price[i]}')
        print('\n총 합계 금액: ' , sum(self.order_price))


    # 지불
    def pay(self):
        while True:
            try:
                p = int(input('\n결제 1.현금, 2.카드'))
            except ValueError:
                print('정수를 입력하세요')
                continue
                
            if p == 1:
                print("직원을 호출하겠습니다.")
                break
            elif p == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("다시 결제를 시도해 주세요")

    # 영수증 데코레이터
    def recpDecorator(func):
        def wrapper(self):
            print('⟝' + '-' * 30 + '⟞')
            for i in range(5):
                print('|' + ' ' * 31 + '|')
            print(f"주문일시: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
            func(self)
            print("\n총 결제 금액: ", sum(self.order_price))
            for i in range(5):
                print('|' + ' ' * 31+ '|')
            print('⟝' + '-' * 30 + '⟞')
        return wrapper
            
    # 주문서 출력 , 데코레이터 지정
    @recpDecorator
    def table(self):
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], " : ",self.order_price[i])
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
