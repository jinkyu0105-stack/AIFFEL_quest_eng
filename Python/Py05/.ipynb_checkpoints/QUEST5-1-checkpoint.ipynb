```python
import random as r

class Account:
    # 객체가 생길 때 (고객 +1) 생성 개수 기록
    account_count = 0
    
    # 생성된 객체를 모아두는 클래스변수 (문자열 아님, 객체 자체)
    accounts = []
    def __init__(self):
        self.bank = "SC은행"
        self.owner = input('계좌 생성 - 예금주 입력: ')
        while True:
            try:
                # 의도적으로 self.balance에 바로 넣지 않는다.
                balance = int(input('계좌 생성 - 초기 잔액: '))
                # 음수 방지
                if balance < 0:
                    print('0원 이상 입력하세요')
                    continue
                # 예외 처리, 검증이 끝난 안전한 값만 '마지막'에 self. 객체에 부여한다.
                self.balance = balance
                break
            except ValueError:
                print('정수를 입력하세요')
                continue

        # zfill(n) : n 자리 수가 될 때까지 왼쪽을 0으로 채운다. 001 부터.. 즉 str() 필수
        num1 = str(r.randint(0,999)).zfill(3)
        num2 = str(r.randint(0,99)).zfill(2)
        num3 = str(r.randint(0,999999)).zfill(6)
        self.account_number = num1 + "-" + num2 + "-" + num3

        # 입,출금 횟수
        self.deposit_count = 0
        self.withdraw_count = 0
        
        # 입,출금 누적 기록 리스트
        self.deposit_log = []
        self.withdraw_log = []

        # 고객이 아닌, 클래스 자체의 변수에 접근하여 총 고객 수 +1
        Account.account_count += 1

        # 시도1. 고객 데이터를 문자열로 만들어 하나의 변수에 담음
        # 시도2. 그 문자열을 accounts 라는 고객 목록 리스트에 append 함.
        # 시도3. 하지만 self 그 자체를 append 하면 고객 객체가 자체를 저장할 수 있음을 알게 됨.
        #client = str(self.bank) + str(self.owner) + str(self.balance) + str(self.account_number)

        # 원래 self 대신 client 였음.
        Account.accounts.append(self)
                                
    # 클래스 메서드란. 
    # 클래스 변수처럼 클래스 용으로 쓰이는 메서드
    # 객체용이 아닌, 클래스가 쓰는 메서드
    # 고객의 역할로 총 고객수를 출력하는 것은 이상하다고 생각.
    # 객체명으로 이 메서드를 출력하는 것이 아닌
    # 클래스명으로 이 메서드를 출력하기 위해서 @classmethod를 부여함.
    # Account.get_account_num() 형태로 사용 가능.
    # 매개변수 cls는 클래스 그 자체를 말한다. 꼭 이 이름으로 해야 하는건 아님
    @classmethod
    def get_account_num(cls):
        print('총 고객 수: ', cls.account_count)


    def deposit(self):
        # 입금 금액 예외 처리
        while True:
            try:
                amount = int(input('입금 금액을 입력하세요: '))
                if amount <= 0:
                    print('1원 이상 입력하세요')
                    continue
            except ValueError:
                print('정수를 입력하세요')
            else:
                self.amount = amount
                self.balance += self.amount
        
                 # 입금 횟수 기록, 입금 금액 누적 기록
                self.deposit_count += 1
                self.deposit_log.append(self.amount)

                # 5회이상 입금시 현재 잔액의 1프로 이자를 더해서 줌
                # 5회째 입금까지 들어간 후에 총 잔액에서 이자를 계산하고 싶어서 밑으로 내렸음.
                # 입금 횟수가 self.deposit_count에 기록 되므로 이 숫자가 5의 배수일때는 매번 5회째 입금이다.
                # 5의 배수는 나눴을때 나머지가 0 이다.
                # 잔액의 1프로를 먼저 계산해서 새 변수에 담고, 잔액에 추가 시킨다.
                if self.deposit_count % 5 == 0:
                    interest = self.balance * 0.01
                    self.balance += interest

            break

        print(f'입금액 {self.amount} : 총 잔액 : {self.balance:,}')
            
    def withdraw(self):
        while True:
            try:
                amount = int(input('출금 금액을 입력하세요: '))
                if amount <= 0:
                    print('1원 이상 입력하세요')
                    continue
                elif amount > self.balance:
                    print('잔액이 부족합니다. 잔액보다 적게 입력하세요')
                    continue                
            except ValueError:
                ('정수를 입력하세요')
                continue
            else:
                self.amount = amount
                self.balance -= self.amount
        
                # 출금 횟수 기록, 출금 금액 누적 기록
                self.withdraw_count += 1
                self.withdraw_log.append(self.amount)

            break

        print(f'출금액 {self.amount} : 총 잔액 : {self.balance:,}')

    def display_info(self):
        return f'은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원'

    def deposit_history(self):
        for i in range(self.deposit_count):
            print(f'입금 : {i+1} \t:\t {self.deposit_log[i]:,}원')

    def withdraw_history(self):
        for i in range(self.withdraw_count):
            print(f'출금 : {i+1} \t:\t {self.withdraw_log[i]:,}원')
```
