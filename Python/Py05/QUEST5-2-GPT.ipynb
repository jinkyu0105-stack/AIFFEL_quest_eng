```python
import random

# =========================
# Q1. Character í´ë˜ìŠ¤
# =========================
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.hp > 0

    def take_damage(self, damage):
        actual_damage = damage - self.defense
        if actual_damage > 0:
            self.hp -= actual_damage

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"{self.name}ì´ {target.name}ì—ê²Œ {damage} ë§Œí¼ ê³µê²©í–ˆë‹¤ ...!")
        target.take_damage(damage)


# =========================
# Q2. Player / Monster í´ë˜ìŠ¤
# =========================
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}ì´ ê²½í—˜ì¹˜ {exp}ì„ ì–»ì—ˆë‹¤! (í˜„ì¬ ê²½í—˜ì¹˜: {self.experience})")

    def level_up(self):
        while self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f"ğŸ‰ ë ˆë²¨ì—…! {self.name}ì˜ ë ˆë²¨ì´ {self.level}ì´ ë˜ì—ˆë‹¤!")


class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)


# =========================
# Q3. battle í•¨ìˆ˜
# =========================
def battle(player, monster):
    print(f"\n{monster.name} ê³¼ì˜ ì „íˆ¬ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤.")

    while player.is_alive() and monster.is_alive():
        # Player ê³µê²©
        player.attack_target(monster)
        print(f"{monster.name}ì˜ ì²´ë ¥: {monster.hp}")

        if not monster.is_alive():
            break

        # Monster ê³µê²©
        monster.attack_target(player)
        print(f"{player.name}ì˜ ì²´ë ¥: {player.hp}")

    if player.is_alive():
        exp = monster.level * 20
        player.gain_experience(exp)
        player.level_up()
        print("ì „íˆ¬ ìŠ¹ë¦¬!")
        print(f"{monster.name} ì„ ì´ê²¼ë‹¤!")
        return True
    else:
        print("ì „íˆ¬ íŒ¨ë°°..")
        return False


# =========================
# Q4. main í•¨ìˆ˜
# =========================
def main():
    monster_dict = {
        'ìŠ¬ë¼ì„': 1,
        'ê³ ë¸”ë¦°': 2,
        'ì˜¤í¬': 3
    }

    player_name = input("í”Œë ˆì´ì–´ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)

        survived = battle(player, monster)
        if not survived:
            print("ê²Œì„ì˜¤ë²„")
            return

    print("\nğŸŠ ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í–ˆìŠµë‹ˆë‹¤! ê²Œì„ í´ë¦¬ì–´!")


if __name__ == "__main__":
    main()
```
