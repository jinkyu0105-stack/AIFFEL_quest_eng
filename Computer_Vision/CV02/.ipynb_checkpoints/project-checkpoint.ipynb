```python
import cv2
import numpy as np
import torch
import torchvision.transforms as T
from torchvision.models.segmentation import DeepLabV3_ResNet101_Weights as W
import matplotlib.pyplot as plt

#  3  
object_paths = ["cat.jpg", "bus.jpg", "motorbike.jpg"]
bg_paths = ["sand.jpg", "grass.jpg", "forest.jpg"]

#  
#  (Eval)
model = deeplabv3_resnet101(weights=W.DEFAULT).eval()
#   ' ( )' 
# torchvision  transforms  Compose  
# Compose:         
transform = T.Compose([
    # OpenCV  (cv2.imread )  numpy   
    # PIL(Python Imaging Library)  .
    # PIL:        "  
    #     PIL   
    T.ToPILImage(),
    # DeepLabV3      520px * 520px
    T.Resize((520, 520)),
    #  PyTorch    
    #   1.  (0~255)  (0.0~1.0)  .
    #   2.   : OpenCV PIL  (, , ) 
    # PyTorch  (, , )  .    .
    T.ToTensor(),
])

# i  (cat.jpg , sand.jpg)
for i, (obj_p, bg_p) in enumerate(zip(object_paths, bg_paths)):
    obj_img = cv2.imread(obj_p)
    bg_img = cv2.imread(bg_p)

    # transform   RGB   .
    #  RGB   RGB     
    obj_img_rgb = cv2.cvtColor(obj_img, cv2.COLOR_BGR2RGB)
    # unsqueeze(0): 3 -> 4 (, , , )
    input_tensor = transform(obj_img_rgb).unsqueeze(0)
    
    # with:      () .  .
    #      
    #      ( )
    #           .)
    with torch.no_grad():
        # !! model   !!
        # ["out"][0] 4   N  3  .
        # output (21, 520, 520)  
        # 21 : DeepLabV3  21 (, ,  ) 
        #   "   "    21
        output = model(input_tensor)["out"][0]
        # argmax(0): 21   , (21, 520, 520)     .
        # (0):  0,  21 . '' . 
        #       (Class ID)   . 
        #   15 '',     15  .
        # 3 2   (21, 520, 520) -> (520, 520)
        # 3 1   (  -> 0: 15: or 8: ....)
        # 1 =  = (, , 1) = (, ) =   *,1*  !
        # 0~255   
        # .byte():  ,    (8) .
        # .cpu():  CPU     ,  GPU    
        # numpy():      
        # cv2.resize np.where        .
        output_predictions = output.argmax(0).byte().cpu().numpy()

    #   Resize
    #    520x520      .
    # cv2.resize(, (, )) 
    # OpenCV   (, , ) (450, 800, 3) -  2 
    # img_orig.shape[0] :  (Height)
    # img_orig.shape[1] :  (Width)
    # INTER_NEAREST (!):    '      '   
    #   ?   15(), 0()  (ID). 
    #     15 0   (7.5  )  
    #      ID   .
    mask_resized = cv2.resize(output_predictions, (obj_img.shape[1], obj_img.shape[0]), 
                              interpolation=cv2.INTER_NEAREST)

    # cv2.applyColorMap: 0~255       (JET) 
    # color_mask     
    # 1 -> 3 
    # mask_resized * 10:  0~21         * 10
    color_mask = cv2.applyColorMap(mask_resized * 10, cv2.COLORMAP_JET)
    
    # (mask_resized > 0) :Numpy   
    #     (Element-wise)  bool 
    # (mask_resized > 0) : 1 True,  False
    # astype(np.uint8): True 1, False 0 .
    # 1 0 255 .  255 or 0 .
    img_mask = (mask_resized > 0).astype(np.uint8) * 255
    # 1 -> 3  :  ->  
    # 255 (255.255.255) 
    # BGR   np.where          .
    img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)

    #       .
    # np.where        .
    bg_resized = cv2.resize(bg_img, (obj_img.shape[1], obj_img.shape[0]))
    # :  (255)  ?
    # True: obj_img 
    # False: bg_resized 
    result_bgr = np.where(img_mask_color == 255, obj_img, bg_resized)

    # 
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1); plt.imshow(obj_img_rgb); plt.title("Original"); plt.axis('off')
    plt.subplot(1, 3, 2); plt.imshow(cv2.cvtColor(color_mask, cv2.COLOR_BGR2RGB)); plt.title("Mask"); plt.axis('off')
    plt.subplot(1, 3, 3); plt.imshow(cv2.cvtColor(result_bgr, cv2.COLOR_BGR2RGB)); plt.title("Result"); plt.axis('off')
    plt.show()
```

```python
cat_path = "cat.jpg"
img_orig = cv2.imread(cat_path)

img_rgb = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)
input_tensor = transform(img_rgb).unsqueeze(0)

with torch.no_grad():
    output = model(input_tensor)["out"][0]
    output_predictions = output.argmax(0).byte().cpu().numpy()

output_predictions_resized = cv2.resize(output_predictions, 
                                        (img_orig.shape[1], img_orig.shape[0]), 
                                        interpolation=cv2.INTER_NEAREST)

seg_map = (output_predictions_resized == 8) 
img_mask = seg_map.astype(np.uint8) * 255
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)

img_bg_mask = cv2.bitwise_not(img_mask_color)

img_orig_blur = cv2.blur(img_orig, (13, 13)) #  

img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)   #   

img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)

plt.figure(figsize=(10, 12))

plt.subplot(2, 1, 1)
plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 1, 2)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.title("Final Out-focusing Result")
plt.axis('off')

plt.tight_layout() #     
plt.show()
```
